```python
## Do not change this but check whether it corresponds with the rows of your dataset. 
labels = [ "HC01", "HC02", "HC03", "HC04", "HC05", "HC06", "HC07", "HC08", "HC09", "HC10", "HC11", "HC12", "HC13", "HC14", "HC15", "HC16", "HC17", "HC18",
  "LC05", "LC06", "LC07", "LC08", "LC09", "LC10", "LC11", "LC12", "LC13", "LC14", "LC15", "LC01", "LC02", "LC03", "LC04",
  "ME13", "ME14", "ME01", "ME02", "ME03", "ME04", "ME05", "ME06", "ME07", "ME08", "ME09", "ME10", "ME11", "ME12"
]
```

```python
#Make sure you have the optimalBBS3004_nometa.csv file in the same directory as this notebook.
from scipy.spatial.distance import pdist, squareform
df = pd.read_csv('optimal_BBS3004_nometa.csv', sep=',', encoding='latin-1')
df.index = labels
df = df.iloc[:, 1:]
df
```

```python
cosine_distances = pdist(df, metric='cosine')
cosine_distances
cosine_distance_matrix = squareform(cosine_distances)
df = pd.DataFrame(cosine_distance_matrix)
df

```

```python
similarities = df.to_numpy()
similarities
euc_observations = squareform(similarities)
euc_observations
```

```python
network_observations, sweep_observations = sd.stad(euc_observations)
```

```python
adjacency_matrix = network_observations.toarray()
adjacency_matrix
```

```python
# collection 'bioMOBS_test_samples' looks like this: you need to run this to match the keys
# [{
#     "_key": "IAM01_1",
#     "_id": "bioMOBS_test_samples/IAM01_1",
#     "_rev": "_dimFSyO---",
#     "individual": "IAM01",
#     "age": 58,
#     "gender": "M",
#     "X__Saturatie": -1.090955188,
#     "Albumine": -0.688442937,
#     "Albumine____": 0.706384552,
#     "Aldosteron": 1.598158214,
#     "Alfa_1_globulinen": -1.139896645,
#     "Alfa_2_globulinen": 0.34075838,
#     ...

aql = "FOR x IN BBS3004_samples RETURN x"
queryResult = db.AQLQuery(aql, rawResults=True, batchSize=100)
data = pd.DataFrame(queryResult)
data
```

```python
#Before running this box make sure you create empty EDGE type named BBS3004_edges (or any other name as log is the name matches with the what you type here) so that you can populate that file by running this script. 
edgesCollection = db["BBS3004_edges"]
edgesCollection.truncate()
docs = []
for i in range(len(np.nonzero(adjacency_matrix)[0])):
    x = np.nonzero(adjacency_matrix)[0][i]
    y = np.nonzero(adjacency_matrix)[1][i]
    x_key = data.iloc[x]["_key"]
    y_key = data.iloc[y]["_key"]

    doc = edgesCollection.createDocument()
    doc['_key'] = x_key + '_' + y_key
    doc['_from'] = "BBS3004_samples/" + x_key
    doc['_to'] = "BBS3004_samples/" + y_key
    doc["weight"] = adjacency_matrix[x][y]
    docs.append(doc)
```

```python
#The crucial part of code do not forget this cell to put it actually in arangODB
edgesCollection.bulkSave(docs) 
```

### proces the sample file properly first so that it's ok for putting into arangodb :

```python
# For modifying the JSON file, upload the modified_file.json into arandob and make sure the connection to that file is defined properly in the obsolete files that annotation inside the interactive stad analyses portal is possible. 
with open("BBS3004_samples.json", "r") as f:
    data = json.load(f)

keys = [
    "HC01", "HC02", "HC03", "HC04", "HC05", "HC06", "HC07", "HC08", "HC09", "HC10", "HC11", "HC12", "HC13", "HC14", "HC15", "HC16", "HC17", "HC18",
  "LC05", "LC06", "LC07", "LC08", "LC09", "LC10", "LC11", "LC12", "LC13", "LC14", "LC15", "LC01", "LC02", "LC03", "LC04",
  "ME13", "ME14", "ME01", "ME02", "ME03", "ME04", "ME05", "ME06", "ME07", "ME08", "ME09", "ME10", "ME11", "ME12"
]


# Modify data

for i, item in enumerate(data):
    # Assign profile
    item["phenotype"] = "control" if i < 18 else "post_covid"
    
    
    # Assign duration
    if i < 18:
        item["phenotype"] = "control"
    elif i < 33:
        item["phenotype"] = "post_covid"
    else:
        item["phenotype"] = "CFS"
    
    # Add _key and individual if keys exist
    if i < len(keys):
        item["_key"] = keys[i]
        item["individual"] = keys[i]

# Save modified JSON file
with open("modified_file.json", "w") as f:
    json.dump(data, f, indent=4)

print("JSON file has been updated and saved as modified_file.json")
```
